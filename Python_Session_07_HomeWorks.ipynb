# 12. Fraction 

```python

def add(x, y):
    res = {}
    if x['d'] == y['d']:
        res['f'] = x['f'] + y['f']
        res['d'] = x['d'] 
    else:
        res['f'] = x['f'] * y['d'] + x['d'] * y['f']
        res['d'] = x['d'] * y['d']
    return res

def sub(x, y):
    res= {}
    if x['d'] == y['d']:
        res['f'] = x['f'] - y['f']
        res['d'] = x['d'] 
    else:
        res['f'] = x['f'] * y['d'] - x['d'] * y['f']
        res['d'] = x['d'] * y['d']
    return res

def mul(x,y):
    res = {}
    res['f'] = x['f'] * y['f']
    res['d'] = x['d'] * y['d']
    return res

def div(x,y):
    res = {}
    res['f'] = x['f'] * y['d']
    res['d'] = x['d'] * y['f']
    return res

def menu():
    print('enter your option:')
    print('1. Summation')
    print('2. Submission')
    print('3. Multiplication')
    print('4. Division')
    a = int(input(''))
    return a
 
 
ch = menu()
s = int(input('form of fraction1: '))
m = int(input('denominator of fraction1: '))
ss = int(input('form of fraction2: '))
mm = int(input('denominator of fraction2: '))

a = {'f':s,'d':m}
b = {'f':ss,'d':mm}

if ch == 1:
    c = add(a, b)
elif ch == 2:
    c = sub(a, b)
elif ch == 3:
    c = mul(a, b)
elif ch == 4:
    c = div(a, b)

print('Result: ',c['f'],'/',c['d'])
```

# 15 Fraction with OO

```python
class fraction:
    def __init__(self, s, m, ss, mm):
        self.s = s
        self.m = m
        self.ss = ss
        self.mm = mm

    def sum(self):
        res = {}
        if self.m == self.mm:
            res['f'] = self.s + self.ss
            res['d'] = self.m 
        else:
            res['f'] = self.s * self.mm + self.m * self.ss
            res['d'] = self.m * self.mm
        return res

    def sub(self):
        res = {}
        if self.m == self.mm:
            res['f'] = self.s - self.ss
            res['d'] = self.m 
        else:
            res['f'] = self.s * self.mm - self.m * self.ss
            res['d'] = self.m * self.mm
        return res

    def mul(self):
        res = {}
        res['f'] = self.s * self.ss
        res['d'] = self.m * self.mm
        return res

    def div(self):
        res = {}
        res['f'] = self.s * self.mm
        res['d'] = self.ss * self.m
        return res

def menu():
    print('enter your option:')
    print('1. sum')
    print('2. sub')
    print('3. mul')
    print('4. div')
    a=int(input(''))
    return a

ch = menu()

s1 = int(input('form of fraction1: '))
m1 = int(input('denominator of fraction1: '))
s2 = int(input('form of fraction2: '))
m2 = int(input('denominator of fraction2: '))

a = fraction(s1, m1, s2, m2)

if ch == 1:
    c = a.sum()
elif ch == 2:
    c = a.sub()
elif ch == 3:
    c = a.mul()
elif ch == 4:
    c= a.div()

print(c['f'],'/',c['d'])
```

# 16 Prime number

```python
def prime(n):
    for i in range(2,n):
        if n % i == 0:
            return False
    else:
        return True
a = int(input('enter num a: '))
if prime(a) == True:
    print(a, 'is prime')
else:
    print(a, 'is not prime')

a = int(input('enter num a: '))
b = int(input('enter num b: '))
for i in range(a,b):
    if prime(i) == True:
        print(i, end=',')
```
