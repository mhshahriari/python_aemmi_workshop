# Store 

```python
import csv 


def read_from_database():
    try:
        products = []
        f = open("database.csv",'r')
        big_text = f.read()
        product_list = big_text.split('\n')

        for i in range(len(product_list)):
            info = product_list[i].split(',')
            #print(info)
            products.append({'id':info[0], 'name':info[1], 'price':info[2], 'count':info[3]})
    except Exception as e:
        print(e)
        products=[]
    return products


def add():
    id = input('enter id: ')
    
    while True:
        for product in products:
            if id == product['id']:
                print('this id is available. try again...')
                id = input('enter id: ')        
        else:   
            name = input('name: ')
            price= input('price: ')
            count= input('count: ')
            
            f = open("database.csv",'a')
            f.write('\n')
            f.write('{},{},{},{}'.format(id , name, price, count))
            products.append({'id':id, 'name':name, 'price':price, 'count':count})
            break
   
    

def search():
    user_input = input('enter a text to search...  ')
    for product in products:
        if product['id'] == user_input or product['name'] == user_input:
            print(product)
            break
    else:
        print('not found')

def edit():
    id = input('enter id for edit: ')
    for product in products:
        if id == product['id'] :
            product['name'] = input('new name :')
            product['price'] = input('new price :')
            product['count'] = input('count :')
    

def remove():
    id = input('enter id for remove: ')
    for product in products:
        if id == product['id'] :
            products.remove(product)

def buy():
    buy_list=[]
    while True:
        id = input('enter id: ')
       
        for product in products:
            if  id == product['id'] :
                count = int(input('enter count: '))
                if count <= int(product['count']):
                    new_count = int(product['count']) - count
                    product['count'] = str(new_count)
                    count = str(count)
                    buy_list.append({'id':product['id'],'name':product['name'],'price':product['price'],'count':count })
                    break
                else:
                    print('this item finished..')
        else:
            print('item not found...try again!')
        flag = input('continue buying ?(y/n)')
        if flag =='n':
            break
   

    #----print factor
    print('\n ---------------Factor------------')
    factor_sum = 0
    for item in buy_list:
        id = item['id']
        name = item['name']
        price = item['price']
        count = item['count']
        item_sum = int(price)*int(count)
        factor_sum += item_sum
        print('{} | {} | {} | {} |{}'.format(id , name, price, count, item_sum))
        print("----------------------------")

    
    print("----------------------------")
    print("Sum of factor: {}".format(factor_sum))
    print("----------------------------")



def show_all():
    products = read_from_database()
    for product in products:
        print(product)

def my_exit():
    flag = input('do you want to save changes?(y/n)')
    if flag =='y':
        f = open("database.csv",'w')
        for product in products:
            f.write('{},{},{},{}'.format(id , name, price, count))
            f.write('\n')
        print('------------')    
        print('save chages was successful')    
        print('------------')    
        exit()


            

def show_menu():
    print('\n*********welcome to store********')
    print('1- add new product')
    print('2- search')
    print('3- edit')
    print('4- remove')
    print('5- buy')
    print('6- show all')
    print('7- exit')

    
    
products = read_from_database()


while(True):
    show_menu()

    choice = input('enter your choice:')

    if choice == '1':
        add()

    elif choice == '2':
        search()

    elif choice == '3':
        edit()

    elif choice == '4':
        remove()

    elif choice == '5':
        buy()

    elif choice == '6':
        show_all()

    elif choice == '7':
        my_exit()


```
