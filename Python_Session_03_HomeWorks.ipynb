# 1 RPS Game

```python
import random
items = ["Rock", "Paper", "Scissor"]

computer_score = 0
user_score = 0
#index =  random.randint(0, 2)

print("0- Rock")
print("1- Paper")
print("2- Scissor")
print("3- Exit")

user_choice_index = int(input("Please enter option#:"))

while (user_choice_index != 3):
    if(user_choice_index not in [0,1,2,3]):
        print("wrong number!!!!")
        user_choice_index = int(input("Enter options 0, 1, 2, 3:"))
        continue
    computer_choice  = random.choice(items)
    user_choice = items[user_choice_index]

    if (computer_choice == "Rock" and user_choice == "Scissor") or (computer_choice == "Scissor" and user_choice == "Paper") or (computer_choice == "Paper" and user_choice == "Rock"):
        computer_score += 1
        print ("Computer Won!. score:" + str(computer_score))
    else:
        user_score += 1
        print ("You Won!. score:" + str(user_score))
    user_choice_index = int(input())

if (computer_score == user_score):
    print("Equal Scores!!!")
elif (computer_score > user_score):
    print("Computer won***. score:" + str(computer_score)) 
else:
     print("You won*** score:" + str(user_score)) 
```

# 5 Check factorial number

```python
num = int(input("Enter your number:"))

i=1
while (True):
    if(num % i == 0):
         num /= i;
    else:
        break
    i +=1
if (num == 1):
    print("Yes")
else:
    print("No")
```

# 6 Check Armstrong numgber

```python
num = input('Enter your number:')
y = 0
c = len(num)
for m in range(c):
    x = int(num[m])**c
    y += x
if y == int(num):
    print("Yes")
else:
    print("No")
```

# 8 Turtle sample

```python
import turtle
import time


p =  turtle.Turtle()
p.shape("turtle") 
p.pencolor ("black")
p.width(1)


for m in range(3, 100):
    num_sides = m
    side_length = 50 
    angle = 360.0 / num_sides

    for i in range(num_sides):        
        p.forward(side_length)
        p.left(angle)
    

turtle.done()
#time.sleep(5)
```
